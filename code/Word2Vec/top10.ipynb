```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

#### 1. Cluster별 Top 10 추천

```python
cluster_df = pd.read_csv('Cluster_df.csv')
```

```python
cluster_df
```

```python
cluster_count_df = cluster_df.groupby('cluster')['item'].count().reset_index().rename(columns={'item':'count'})
```

```python
cluster_count_df
```

```python
raw_df = pd.read_csv('../../data/train/train_ratings.csv')[['user','item']]
```

```python
raw_df
```

```python
df = raw_df.merge(cluster_df, on='item')
```

```python
df
```

```python
item_df = cluster_df.merge(df.groupby('item')['user'].nunique().reset_index(), on='item').rename(columns={'user':'count'})
```

```python
item_df
```

```python
group_df = df.groupby(['user','cluster'])['item'].count().unstack().T.fillna(0)
```

```python
group_df
```

```python
for cluster in cluster_count_df.index.values:
    group_df.loc[cluster] = group_df.loc[cluster].apply(lambda r: r/cluster_count_df.loc[cluster,'count'])
```

```python
group_df
```

```python
user_series = raw_df['user'].unique()

first = []
second = []
third = []
fourth = []
fifth = []
sixth = []
for user in tqdm(user_series):
    clusters = group_df[user].argsort()[::-1].values
    first.append(clusters[0])
    second.append(clusters[1])
    third.append(clusters[2])
    fourth.append(clusters[3])
    fifth.append(clusters[4])
    sixth.append(clusters[5])
    
```

```python
user_df = pd.DataFrame(zip(user_series,first,second, third, fourth, fifth, sixth), columns=('user','cluster1','cluster2', 'cluster3', 'cluster4', 'cluster5', 'cluster6'))
user_df.set_index('user', inplace=True)
```

```python
user_df
```

```python
item_df.sort_values(['cluster','count'], inplace=True, ascending=False)
```

```python
item_df
```

```python
from tqdm import tqdm

sub_item = np.array([])
for user in tqdm(raw_df['user'].unique()):
    i = 1
    cluster = user_df.loc[user][f'cluster{i}']
    temp_cluster_df = item_df[item_df['cluster']==cluster]
    top10 = temp_cluster_df[~temp_cluster_df['item'].isin(raw_df[raw_df['user'] == user]['item'].values)][:10]['item'].values
    diff = 10 - len(top10)
    while diff:
        i += 1
        cluster = user_df.loc[user][f'cluster{i}']
        temp_cluster_df = item_df[item_df['cluster']==cluster]
        top_n = temp_cluster_df[~temp_cluster_df['item'].isin(raw_df[raw_df['user'] == user]['item'].values)][:diff]['item'].values
        top10 = np.concatenate((top10, top_n))
        diff = 10 - len(top10)
    sub_item = np.concatenate((sub_item, top10))
```

```python
print(sub_item)
print(len(sub_item))
```

```python
item_df[item_df['cluster'] == 2][:20]
```

```python
sub_user = user_df.index.values.repeat(10)
```

```python
print(sub_user)
print(len(sub_user))
```

```python
sub_df = pd.DataFrame(zip(sub_user,sub_item), columns=['user','item'])
sub_df['item'] = sub_df['item'].apply(lambda x: int(x))
sub_df.to_csv("output/cluster_normalized_ranking_submission.csv", index=False)
```

#### 2. 유저가 본 영화의 평균 임베딩과 가장 가까운 10개

```python
emb_df = pd.read_csv('Word2Vec_not_shuffle_emb_df.csv')
```

```python
emb_df
```

```python
raw_df = pd.read_csv('../../data/train/custom_train_ratings.csv')[['user','item']]
```

```python
merged_df = raw_df.merge(emb_df, on='item').sort_values('user')
```

```python
merged_df
```

```python
group_df = merged_df.groupby('user')[[str(x) for x in range(500)]].mean()
```

```python
group_df
```

```python
from tqdm import tqdm
import torch
from torch.linalg import norm
import heapq

emb_tensor = torch.tensor(emb_df.iloc[:,:-1].values)
group_tensor = torch.tensor(group_df.values)
user2idx = {u:i for i,u in enumerate(group_df.index.values)}

item_list = []
seen_count = 0
for user in tqdm(group_df.index.values):
    idx = user2idx[user]
    emb = group_tensor[idx]
    heap = []
    for i in range(6807):
        sim = emb.matmul(emb_tensor[i]).item()
        heapq.heappush(heap, (-sim,emb_df.iloc[i,-1]))
        #dist = norm(emb - emb_tensor[i])
        #heapq.heappush(heap, (dist,emb_df.iloc[i,-1]))
    
    top10 = []
    count = 0
    seen = raw_df[raw_df['user'] == user]['item'].values
    while heapq and count < 10:
        top = heapq.heappop(heap)[1]
        if top in seen:
            seen_count += 1
            continue
        top10.append(top)
        count += 1
    item_list += top10
```

```python
print(len(item_list))
```

```python
print(f"이미 관람하여 제외된 영화 수: {seen_count}")
```

```python
sub_user = raw_df['user'].unique().repeat(10)
```

```python
print(sub_user)
print(len(sub_user))
```

```python
sub_df = pd.DataFrame(zip(sub_user,item_list), columns=['user','item'])
sub_df.to_csv("output/custom_avg_500_emb_matmul_submission.csv", index=False)
```

```python
concat_df = pd.concat((raw_df, sub_df), axis=0)
```

```python
len(concat_df)
```

```python
drop_df = concat_df.drop_duplicates(('user','item'))
```

#### 3. Cluster별 Voting Top 10 추천

```python
cluster_df = pd.read_csv('../IBCF/IBCF_cluster.csv')
```

```python
cluster_df
```

```python
cluster_count_df = cluster_df.groupby('cluster')['item'].count().reset_index().rename(columns={'item':'count'})
```

```python
cluster_count_df
```

```python
raw_df = pd.read_csv('../../data/train/custom_train_ratings.csv')[['user','item']]
```

```python
raw_df
```

```python
df = raw_df.merge(cluster_df, on='item')
```

```python
df
```

```python
item_df = cluster_df.merge(df.groupby('item')['user'].nunique().reset_index(), on='item').rename(columns={'user':'count'})
```

```python
item_df
```

```python
group_df = df.groupby(['user','cluster'])['item'].count().unstack().T.fillna(0)
```

```python
group_df
```

```python
for cluster in cluster_count_df.index.values:
    group_df.loc[cluster] = group_df.loc[cluster].apply(lambda r: r/cluster_count_df.loc[cluster,'count'])
```

```python
group_df
```

```python
item_df.sort_values(['cluster','count'], inplace=True, ascending=False)
```

```python
item_df
```

```python
n_cluster = len(group_df.index)
item_df_list = [item_df[item_df['cluster']==cluster] for cluster in range(n_cluster)]
```

```python
np.random.seed(42)

sub_item = np.array([])
for user in tqdm(raw_df['user'].unique()):
    top10 = []
    preference = group_df.loc[:,user]
    preference /= np.sum(preference)

    seen_item = raw_df[raw_df['user'] == user]['item'].values
    while len(top10) < 10:
        cluster = np.random.choice(n_cluster, p=preference)
        for i in item_df_list[cluster]['item'].values:
            if (not i in top10) and (not i in seen_item):
                top10.append(i)
                break
    sub_item = np.concatenate((sub_item, top10))
```

```python
print(sub_item)
print(len(sub_item))
```

```python
sub_user = raw_df.user.unique().repeat(10)
```

```python
print(sub_user)
print(len(sub_user))
```

```python
sub_df = pd.DataFrame(zip(sub_user,sub_item), columns=['user','item'])
sub_df['item'] = sub_df['item'].apply(lambda x: int(x))
sub_df.to_csv("output/IBCF_cluster_voting_submission.csv", index=False)
```

```python
emb_df = pd.read_csv('new_Word2Vec_not_shuffle_emb_df.csv')
```

```python
emb_df = emb_df.set_index('item', drop=True).sort_index()
```

```python
emb_df
```

```python
emb_norm = emb_df / np.linalg.norm(emb_df, axis = 1, keepdims=True)
```

```python
emb_norm
```

```python
emb_sim = emb_norm @ emb_norm.T
```

```python
emb_sim
```

```python
raw_df = pd.read_csv('../../data/train/custom_train_ratings.csv')[['user','item']]
```

```python
raw_df['label'] = [1]*len(raw_df)
```

```python
pivot = pd.pivot(raw_df, index='user', columns='item', values='label').fillna(0)
```

```python
pivot
```

```python
result_matrix = pivot @ emb_sim
```

```python
mask = pivot * (-np.max(result_matrix.values))
```

```python
result_matrix += mask
```

```python
result_matrix
```

```python
result_matrix.isna().values.any()
```

```python
top10_item_idx = np.argsort(result_matrix, axis=1).iloc[:, :-11:-1]
```
